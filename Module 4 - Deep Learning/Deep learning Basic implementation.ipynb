Before moving onto deep learning lets visualize the sigmoid function which we were using and talking about for such a long tim.  

```python
# importing the libraries
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# It will generate 100 random values between the numbers -10 and 10
input = np.linspace(-20,20,41)
print(input)
```

```python
# define the sigmoid function 
def sigmoid(x):
    return (1/(1+np.exp(-x)))
```

```python
output = sigmoid(input)
plt.plot(input,output)
plt.xlabel('Input')
plt.ylabel('output')
plt.title('This is how the sigmoid function looks like')
```

## Simple implementation of neural network in python. 


#### Lets imagine we have basically two input features that is x1 and x2

```python
# prepare the input feature
import pandas as pd
input_features = np.array([[1,2],[3,4],[2,6],[0,2]])
print(input_features)
input_features.shape
```

```python
# Prepare the output feature
target_output = np.array([[0,1,1,0]])
target_output = target_output.reshape(4,1)

```

```python
# initialize the weights
weights = np.random.randn(2,1)
weights.shape
print(weights)
```

```python
# Initializing the bias and learning rate
bias = 1 
lr = 0.1
```

```python
# defining the function for the derivative of the sigmoid function
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(10000):
    inputs = input_features
    # Forward propagation
    z = np.dot(inputs,weights) + bias
    a = sigmoid(z)
    error = target_output - a
    x = error.sum
    print(x)
    deriv = error * sigmoid_der(a)
    inputs = input_features.T
    derivative = np.dot(inputs,deriv)
    weights -= lr*derivative
    for i in deriv:
        bias -= lr*i
    
```

```python
print(weights)
print(bias)
```

```python
test_point = np.array([1,10])
result1 = np.dot(test_point,weights) + bias
result2 = sigmoid(result1)
print(result2)
```
