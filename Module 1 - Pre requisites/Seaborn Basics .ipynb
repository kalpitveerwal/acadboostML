When we are making analysis using only one input variable, then it is called univariate analysis

Seaborn also helps to make analysis on more than two features

### For Numerical Variable

#### Correlation with Heatmap

So First what is correlation ? 
- Correlation is a technique used to show how two features are related. Correlation can be found out if the variable is integer or float.
Where as if the variable are categorical feature that times the correlation can not be found out  

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

#### JointPlot 
Helps to do univariate analysis mostly but is used in bi-vairate analysis as well 

```python
sns.jointplot(x='tip', y = 'total_bill', data= df , kind = 'hex')
```

```python
sns.jointplot(x='tip', y = 'total_bill', data= df , kind = 'reg')
```

#### PairPlot 

By default, this function will create a grid of Axes such that each numeric variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

#### Distplot 
Flexibly plot a univariate distribution of observations.

```python
sns.distplot(df['tip'])
```

### For Categorical Variable 
Categorical Data is the data that generally takes a limited number of possible values. Also, the data in the category need not be numerical, it can be textual in nature

#### Distribution plots 
1. countPlot 
2. BoxPlot 
3. ViolinPlot 
4. BarPlot 

#### CountPlot

```python
sns.countplot('day',data=df)
```

```python
sns.barplot(y='total_bill',x = 'smoker',data=df)
```

#### Box Plot 
 boxplot is a method for graphically depicting groups of numerical data through their quartiles.

```python
sns.boxplot('smoker','total_bill',data = df)
```

#### Violin Plot

 It is similar to a box plot, with the addition of a rotated kernel density plot on each side

```python
sns.violinplot(x='total_bill', y='day', data = df, palette = 'rainbow')
```
